## One-hot parse for zipcodes:

```python
X_train = pd.get_dummies(X_train, columns=['zipcode'])
X_test = pd.get_dummies(X_test, columns=['zipcode'])
```

## Transform data:

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

def take_year(house):
    return (int(house.split('-')[0]) - 2014) * 365 + int(house.split('-')[1]) * 30 + int(house.split('-')[2])

def increase(x):
    return int(x) ** 2

def parse(data, features = 8):
    data['waterfront'] = data['waterfront'] * 10
    data['long'] = np.abs(data['long'] + 122.21300580444336)
    data['lat'] = np.abs(data['lat'] - 47.62979889)
    data['yr_built'] = np.abs(data['yr_built'] - 1905)
    data['date'] = data['date'].apply(take_year)
    
    data['sqft_living2'] = np.sqrt(data['sqft_living'])
    data['sqft_lot2'] = np.sqrt(data['sqft_lot'])
    data['sqft_above2'] = np.sqrt(data['sqft_above'])
    data['sqft_basement2'] = np.sqrt(data['sqft_basement'])
    
    data['floors2'] = data['floors'] ** 2
    data['bedrooms2'] = data['bedrooms'] ** 2
    data['grade2'] = data['grade'] ** 2
    data['yrs'] = data['yr_built'] + data['yr_renovated']

parse(X_train)
parse(X_test)
```

## Train our modelUse polinomial features and regression for learning

```python
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression, LassoLarsCV, BayesianRidge, ElasticNet, HuberRegressor, ElasticNetCV, SGDRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.preprocessing import StandardScaler

def mape(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

steps = 10
sum_err = 0
poly = preprocessing.PolynomialFeatures(degree = 2)
X_train = poly.fit_transform(X_train)
ss = StandardScaler()
for i in range(steps):
    x1_train, x1_test, y1_train, y1_test = train_test_split(X_train, y_train, train_size = 0.75)
    x1_train = ss.fit_transform(x1_train)
    x1_test = ss.fit_transform(x1_test)
    y1_train = np.log(y1_train)
    LR = HuberRegressor(alpha=1e-3).fit(x1_train, y1_train)
    res = LR.predict(x1_test)
    res = np.exp(res)
    sum_err += mape(res, y1_test)
    print(i)
print(sum_err / steps)
```

## Save result

```python
def saver(path, res):
    with open(path, 'w') as out:
        print('Id,Price', file = out)
        for i in range(len(res)):
            print('{i},{res}'.format(i = i + 1, res=res[i]), file = out)

ss = StandardScaler()
poly = preprocessing.PolynomialFeatures(degree = 2)

X_train = poly.fit_transform(X_train)
X_train = ss.fit_transform(X_train)
X_test = poly.fit_transform(X_test)
X_test = ss.fit_transform(X_test)

y_train = np.log(y_train)
LR = ElasticNet().fit(X_train, y_train)
res = LR.predict(X_test)
res = np.exp(res)
saver('res.csv', res)
```
